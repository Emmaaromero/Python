```python
# librería

import matplotlib.pyplot as plt

import numpy as np
```

```python
#vamos a generar datos con linspace. Se le pasan 3 argumentos: inicio, fin y cantidad de valores
x=np.linspace(0,10,50)
x
# la fucnión "linspace" genera un array NumPy formado por "n" números equiespaciados entre dos numeros dados
```

```python
#ahora vamos a hacer una sinusoide y una cosinusoide
y1=np.sin(x)
print(y1)
y2=np.cos(x)
print(y2)
```

```python
#ahora graficamos el seno (fijense que hay que poner  plt.show para poder ver el grafico)
plt.plot(x,y1)
plt.xlabel("x") #ponemos etiqueta al eje x
plt.ylabel("seno(x)")  #ponemos etiqueta al eje y
plt.title("Seno(x)")  #ponemos título al gráfico
plt.show() #mostramos el gráfico
```

```python
#ahora graficamos el coseno (ponegan ustedes etiquetas y titulo)
plt.plot(x,y2)
plt.xlabel("x") #ponemos etiqueta al eje x
plt.ylabel("cos(x)")  #ponemos etiqueta al eje y
plt.title("cos(x)")  #ponemos título al gráfico
plt.show() #mostramos el gráfico
plt.show()
```

```python
#si queriamos hacer una recta podríamos haber puesto la función y graficarla

y3=3*x+0.5
plt.plot(x,y3)
plt.show()
```

```python
#si queriamos hacer otra función por ejemplo una parábola podríamos haber puesto la función y graficarla
y4=-2*x**2+5*x-3
plt.plot(x,y4)
plt.show()
```

```python
#ahora vamos a hacer un scatterplot con los datos de los alquileres de bicicletas de acuerdo con temperatura en diferentes fechas (el dato ha sido normalizado)
temp=pd.read_csv("tempday.csv")
temp.head()
```

```python
plt.scatter('temp', 'cnt', data=temp, c='season') 
#graficamos la columna temp en el eje x y cnt en el eje y, del dataset temp y los puntos los pintamos de color de acuerdo a la columna season
plt.xlabel('Normalized Temperature', fontsize='large')
plt.ylabel('Count of Total Bike Rentals', fontsize='large')
```

```python
plt.hist(temp.temp,bins=10) #bins: cantidad de intervalos
plt.rcParams['figure.figsize']=[7,7]
plt.show
#vamos a ver que no son normales
```
