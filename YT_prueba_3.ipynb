```python
from googleapiclient.discovery import build #crea un Recurso para interectuar con la API
#devuelve un objeto con metodos dentro para interactuar con la API
# Documentación de la librería: https://googleapis.github.io/google-api-python-client/docs/epy/googleapiclient.discovery-module.html 

api_key ="**********************************"
video_id= "1BpTBzDQuRE"
```

```python
# Se construye un Recurso para la API de YT
resource = build(
    serviceName = 'youtube', #nombre del servicio
    version = "v3", # Versión del servicio
    developerKey= api_key # key obtenida en  https://code.google.com/apis/console
)
# devuelve un objeto con metodos que usamos para interactuar con la API
```

```python
resource
```

```python
# Crear una petición que traiga 100 comentarios
request = resource. commentThreads().list( #utilizamos la función que trae comentarios y los convertimos a lista
                            part="snippet", #nos traemos solo el snippet
                            videoId=video_id,
                            maxResults= 100,   
                            order="orderUnspecified")  #top comments.
```

```python
request
```

```python
# Ejecutamos la petición
response =request.execute()
```

```python
response
```

```python
# Traerse los primeros 50 comentarios
items = response["items"][:50]
```

```python
print("------------------------------------------------------------------------------------------------------")
for item in items:
    item_info = item["snippet"]
    
    #the top level comment can have sub reply comments
    topLevelComment = item_info["topLevelComment"]
    comment_info = topLevelComment["snippet"]
    
    print("Comment By:", comment_info["authorDisplayName"])
    print("Comment Text:" ,comment_info["textDisplay"])
    print("Likes on Comment:", comment_info["likeCount"])
    print("Comment Date: ", comment_info['publishedAt'])
    print("================================\n")
```

```python
from googleapiclient.discovery import build

#credentials 
api_key ="****************************************"
video_id= "1BpTBzDQuRE"

#build a resource for youtube
resource = build('youtube', 'v3', developerKey=api_key)

#create a request to get 20 comments on the video
request = resource. commentThreads().list(
                            part="snippet",
                            videoId=video_id,
                            maxResults= 20,   #get 20 comments
                            order="orderUnspecified")  #top comments.
#execute the request
response =request.execute()

#get first 10 items for from 20 comments 
items = response["items"][:10]

print("------------------------------------------------------------------------------------------------------")
for item in items:
    item_info = item["snippet"]
    
    #the top level comment can have sub reply comments
    topLevelComment = item_info["topLevelComment"]
    comment_info = topLevelComment["snippet"]
    
    print("Comment By:", comment_info["authorDisplayName"])
    print("Coment Text:" ,comment_info["textDisplay"])
    print("Likes on Comment :", comment_info["likeCount"])
    print("Comment Date: ", comment_info['publishedAt'])
    print("================================\n")
```
