```python
#vamos a usar una lista para importar los datos de un csv, inicializamos la lista
datos=[]
#creamos una lista vacia
print(type(datos))
print('datos: ', datos)
```

```python
#vamos a recorrer el archivo datos_pandas.csv linea a linea con un for. 

#El archivo conviene ponerlo donde tienen la carpeta de trabajo del Jupyter Notebook

#presten especial atención a la indentación de las líneas de código dentro del for

#En Python se usa la indentación para indicar si una línea esta dentro o fuera de un bloque. 

#Se usa en los #bloques for, while, if etc. Si no quiero que la línea este dentro del bloque la "desindento"

for i in open("datos_pandas.csv"): # para el parametro cuando abramos "datos_pandas.csv"

    fila=i.split(',') #los datos de cada fila se van a importar como textos

    dato=list(map(float,fila)) #ahora se convierten a numeros (en python 2 es dato=map(float,fila)

    datos.append(dato) #los agregamos a la lista
```

```python
#Pueden probar imprimiendo dentro del bloque (se va a mostrar cada dato que se está leyendo):
for i in open("datos_pandas.csv"):
  #para la variable "i" en el elemento iterable "datos_pandas.csv"

    fila=i.split(',') #los datos de cada fila se van a importar como textos
    # se crea una lista denominada "fila" que contendrá: 
    # la función "split()" devuelve una lista de caracteres de los elementos sepradas según el parámetro pasado dentro de la función
    # en este caso el aprametro pasado sería (",")
    # recordemos que hasta ahora el archivo csv no fue imoprtado a una estructura que lo reconozca como filas y columnas o matricialmente
    # por lo tanto poseemos un archivo plano delimitado por comas
    # entonces luego de esta función obtendremos una lista unica, con todos los datos del csv separados y no de forma plano como en el archivo
    print('\n fila: ', fila)

    dato=list(map(float,fila)) # ahora se convierten a numeros (en python 2 es dato=map(float,fila)
    #convertimos el tipo caracter de la función anterior a float (que estaban contenidos en la lista denominada "fila")
    # La función map() toma una función y una lista y aplica esa función a cada elemento de esa lista, produciendo una nueva lista.
    # vamos a obtener una nueva lista denominada "dato"  que contendra lo mismo que "fila", pero con datos tipo FLOAT
    print('\n dato:', dato)

    datos.append(dato) #los agregamos a la lista
    # La función append() agrega un ítem al final de la lista
    print('\n dato:', dato)
    # en este caso agregamos a la lista "datos", los valores de la lista "dato"
    # recordemos que la lista denominda "datos" la creamos vacia
```

```python
#ahora vamos a crear dataframes con los datos a mano, a partir de listas
usuarios=[1,2,3]
nombres=["juan","maria","luis"]
```

```python
print(nombres, usuarios)
```

```python
data=pd.DataFrame(list(zip(usuarios, #la función zi: es utilizada para mapear los mismos índices en más de un iterable.
                           nombres)),#entonces lo que hara es a las varaibles 'usuarios' y 'nombres' les asignará el mismo indice
                  columns=['usuarios',
                           'nombres'])                    
data
```

```python
#vemos que el tipo es correcto
type(data)
```

```python
#ahora crearemos otro con los mismos Ids pero otros datos
Ids=[1,2,1,1,1,3,3,2,1,2,2,2]
publicidad=[101,105,108,400,203,198,209,115,274,222,734,219]
```

```python
#Ahora deberas crear un nuevo dataframe con estos datos y llamarlo ads, las columnas seran usuarios, ad_id
ads=pd.DataFrame(list(zip(Ids,
                          publicidad)),
                 columns=['usuarios',
                          'publicidades']) 
ads
```

```python
#lo que debemos hacer ahora es un "join" entre ambas listas

df=pd.merge(data,
            ads, 
            on='usuarios')
```

```python
df #vemos que justamente joineó por el camo usuarios y quedó la data en un solo dataframe
```
